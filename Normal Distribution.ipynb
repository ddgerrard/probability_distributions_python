Examples:

<h2><center><u>Normal Distribution Notation</u></center></h2>
\begin{equation*}
\Large
N(\mu,\sigma^2)
\end{equation*}

<h2><center><u>Normal p.d.f.</u></center></h2>
\begin{equation*}
\Large
\phi(X; \mu, \sigma^2) = \frac{1}{\sigma\sqrt{2\pi}}e^\frac{-(X-\mu)^2}{2\sigma^2}
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.special import erf, erfinv
```

# Normal Distribution using scipy

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
mu = 5000
sigma = 100
x = np.linspace(norm.ppf(0.001, loc=mu, scale=sigma), norm.ppf(0.999, loc=mu, scale=sigma), 1000) # ppf = percent point function
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(221)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma)) # pdf = probability distribution function
plt.title('pdf')
plt.subplot(222)
plt.plot(x, norm.cdf(x, loc=mu, scale=sigma)) # cdf = cumumlative distribution function
plt.title('cdf')
plt.subplot(223)
r = norm.rvs(loc=mu, scale=sigma, size=10000) # rvs = random variates
plt.hist(r, 100, density=True)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma))
plt.title('hist')
r = norm.rvs(loc=mu, scale=sigma, size=100000)
plt.subplot(224)
q = np.linspace(0.001, 0.999, 1000)
plt.plot(q, norm.ppf(q, loc=mu, scale=sigma))
plt.title('quantile')
plt.show()
```

# Normal Distribution explicitly in Python

```python
def normalDist(mu, sigma):
    Xs = np.linspace(mu-4*sigma,mu+4*sigma,101)
    Ys = 1/np.sqrt(2*np.pi*pow(sigma,2))*np.exp(-pow((Xs-mu),2)/(2*pow(sigma,2)))
    CDFs = 1/2*(1+erf((Xs - mu)/(sigma*np.sqrt(2))))
    return Xs, Ys, CDFs
```

# pdf - probability distribution function

```python
mu, sigma = 0, 1 # mean and standard deviation
Xs, Ys, CDFs = normalDist(mu, sigma)
```

```python
plt.plot(Xs,Ys)
plt.show()
```

```python
A = np.random.normal(mu, sigma,10000)
plt.hist(A,50, density=True)
plt.plot(Xs, Ys, linewidth=2, color='r')
```

# cdf - cumulative distribution function

```python
mu, sigma = 0, 1 # mean and standard deviation
Xs, Ys, CDFs = normalDist(mu, sigma)
```

```python
plt.plot(Xs,CDFs)
plt.show()
```

# Quantile

```python
Fs = np.linspace(0,1,100001)
Qs = mu + sigma*np.sqrt(2)*erfinv(2*Fs - 1)
plt.plot(Fs,Qs)
plt.show()
```
